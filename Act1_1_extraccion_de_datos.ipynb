Se muestran los elementos de la columna de 'PRECIO_UNITARIO' mayor a 1000 y los elementos de 'FECHA_DOC' que coincidan con '2022-01-05'

```python
filtro8.to_csv("Filtro8.csv")

from google.colab import files

files.download("Filtro8.csv")
```

Muestra elementos de la columna 'PRECIO_UNITARIO' menor a 100 o los elementos de la columa 'FECHA_DOC' que coinciden con '2022-01-05'

```python
filtro9 = df[(df["PRECIO_UNITARIO"] < 100) | (df["FECHA_DOC"] == "2022-01-05")]
filtro9
```

```python
filtro9.to_csv("Filtro9.csv")

from google.colab import files

files.download("Filtro9.csv")
```

No va a mostrar los elementos que sean mayores a 450 y que tengan la fecha '2022-01-05'

```python
filtro10 = df[(df["PRECIO_UNITARIO"] < 450) & ~(df["FECHA_DOC"] == "2022-01-05")]
filtro10
```
